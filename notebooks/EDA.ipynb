# ONTIMEPROPHET

## GOAL: To predict time taken for food delivery.

## Introduction About Dataset:
   

### There are 20 independent variables (including id and delivery person id)

*Delivery_person_ID : Unique ID to identify each delivery person.

*Delivery_person_Age: Specifies age of delivery person.

*Delivery_person_Ratings: Specifies the performance of deliver.

*Restaurant_latitude:Latitude of Restaurant for location of restuarant.

*Restaurant_longitude:Longitude of Restaurant for location of restuarant.

*Delivery_location_latitude:Latitude of Delivery location.

*Delivery_location_longitude:Longitude of Delivery location.

*Order_Date:This refers to the date when a customer places an order.

*Time_Orderd:This refers to the moment when a customer places an order.

*Time_Order_picked:The timestamp indicating when the order is picked up for delivery.

*Weather_conditions: The prevailing weather conditions during the delivery process.

*Road_traffic_density:The level of traffic congestion on the roads during transportation.

*Vehicle_condition:The state or condition of the delivery vehicle.

*Type_of_order:The category or nature of the order being delivered.

*Type_of_vehicle:The specific kind of vehicle used for transportation.

*multiple_deliveries:Whether the delivery involves multiple packages or stops.

*Festival: The presence or influence of a festival during the delivery timeframe.

*City:The location or city where the delivery is taking place.

*Time_taken(min):The duration, in minutes, taken to complete the delivery process.

```python
Weather_conditions_map={'Fog':6, 'Stormy':5, 'Sandstorms':4, 'Windy':3, 'Cloudy':2, 'Sunny':1}
Road_traffic_density_map={'Jam':4,'High':3,'Medium':2,'Low':1}
Type_of_order_map={'Snack':2,'Meal':3,'Drinks':1,'Buffet':4}
Type_of_vehicle_map={'motorcycle':4,'scooter':3,'electric_scooter':2,'bicycle':1}
Festival_map={'Yes':1,'No':0}
City_map={'Metropolitian':3,'Urban':2,'Semi-Urban':1}
```

```python
df['Weather_conditions']=df['Weather_conditions'].map(Weather_conditions_map)
df['Road_traffic_density']=df['Road_traffic_density'].map(Road_traffic_density_map)
df['Type_of_order']=df['Type_of_order'].map(Type_of_order_map)
df['Type_of_vehicle']=df['Type_of_vehicle'].map(Type_of_vehicle_map)
df['Festival']=df['Festival'].map(Festival_map)
df['City']=df['City'].map(City_map)
```
